```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from ultralytics import YOLO

# Load your trained YOLOv8 model (change path to your best.pt)
model = YOLO('/content/gdrive/My Drive/add/runs/detect/train/weights/best.pt')

val_images_path = '/content/gdrive/My Drive/add/yolo_formatted_dataset/images/val'
val_labels_path = '/content/gdrive/My Drive/add/yolo_formatted_dataset/labels/val'

def build_confusion_matrix(val_images_path, val_labels_path, class_names):
    y_true = []
    y_pred = []
    label_files = [f for f in os.listdir(val_labels_path) if f.endswith('.txt')]

    # Add an extra class for "No Detection"
    extended_class_names = list(class_names) + ['NoDetect']
    no_detect_class = len(extended_class_names) - 1  # index for NoDetect

    for label_file in label_files:
        img_file = os.path.splitext(label_file)[0]

        # Find corresponding image file with common extensions
        possible_img_extensions = ['.jpg', '.jpeg', '.png']
        img_path = None
        for ext in possible_img_extensions:
            candidate = os.path.join(val_images_path, img_file + ext)
            if os.path.exists(candidate):
                img_path = candidate
                break
        if img_path is None:
            print(f"Warning: No image found for label {label_file}")
            continue

        # Read ground truth classes
        with open(os.path.join(val_labels_path, label_file), 'r') as f:
            true_classes = [int(line.split()[0]) for line in f.readlines()]

        # Run model prediction on image
        results = model(img_path)
        pred_classes = []
        for r in results:
            boxes = r.boxes
            for box in boxes:
                conf = box.conf.item()
                if conf >= 0.25:  # confidence threshold
                    pred_classes.append(int(box.cls.item()))

        # Handle no detection or mismatch by assigning 'NoDetect' for all true labels
        if len(true_classes) == len(pred_classes) and len(true_classes) > 0:
            y_true.extend(true_classes)
            y_pred.extend(pred_classes)
        else:
            y_true.extend(true_classes)
            y_pred.extend([no_detect_class] * len(true_classes))

    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(extended_class_names))))

    return cm, extended_class_names

def plot_confusion_matrix(cm, class_names):
    fig, ax = plt.subplots(figsize=(10, 10))
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(
        xticks=np.arange(cm.shape[1]),
        yticks=np.arange(cm.shape[0]),
        xticklabels=class_names,
        yticklabels=class_names,
        ylabel='True label',
        xlabel='Predicted label',
        title='Confusion Matrix with NoDetect Class',
    )

    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(
                j,
                i,
                format(cm[i, j], 'd'),
                ha="center",
                va="center",
                color="white" if cm[i, j] > thresh else "black",
            )
    fig.tight_layout()
    plt.show()

# Build and plot confusion matrix
cm, extended_class_names = build_confusion_matrix(val_images_path, val_labels_path, model.names)
plot_confusion_matrix(cm, extended_class_names)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load model
model = YOLO('runs/detect/train/weights/best.pt')

# Run validation
metrics = model.val()

# Get full class name dictionary (e.g., 0: "English", 1: "Hindi", ...)
class_names = model.names

# Initialize all 10 classes
full_class_list = ["Bengali", "English", "Gujarati", "Hindi", "Kannada",
                   "Malayalam", "Numeral", "Punjabi", "Tamil", "Telugu"]

# Create a mapping from name to ID
name_to_id = {v: k for k, v in class_names.items()}

# Get actual mAP@0.5 per-class from validation
if hasattr(metrics.box, 'map_per_class'):
    per_class_map = metrics.box.map_per_class  # numpy array of length = num of classes seen in val
else:
    per_class_map = [0] * len(class_names)  # fallback

# Build aligned mAP list for all 10 classes
map50_scores = []
for name in full_class_list:
    if name in name_to_id:
        idx = name_to_id[name]
        if idx < len(per_class_map):
            map50_scores.append(float(per_class_map[idx]))
        else:
            map50_scores.append(0.0)
    else:
        map50_scores.append(0.0)

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(full_class_list, map50_scores, color='skyblue')
plt.xlabel('Text Classes')
plt.ylabel('mAP@0.5')
plt.title('Per-Class mAP@0.5 for Text Detection')
plt.ylim(0, 1)
plt.xticks(rotation=45, ha='right')

# Annotate bars with scores
for bar, score in zip(bars, map50_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             f"{score:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import glob

# Ground truth and prediction directories
gt_dir = os.path.join(ROOT_DIR, 'labels/val')
pred_dir = '/content/runs/detect/val/labels/'

# Get list of ground truth label files
gt_files = sorted(glob.glob(gt_dir + '/*.txt'))

# Initialize counters
total_labels = 0
correct_predictions = 0

for gt_file in gt_files:
    # Corresponding prediction file
    pred_file = os.path.join(pred_dir, os.path.basename(gt_file))

    if not os.path.exists(pred_file):
        continue  # If prediction file doesn't exist, skip

    # Read ground truth class IDs
    with open(gt_file, 'r') as f:
        gt_lines = f.readlines()
    gt_classes = [line.strip().split()[0] for line in gt_lines]

    # Read predicted class IDs
    with open(pred_file, 'r') as f:
        pred_lines = f.readlines()
    pred_classes = [line.strip().split()[0] for line in pred_lines]

    # Compare class IDs one by one
    total_labels += len(gt_classes)
    correct_predictions += sum(1 for gt, pred in zip(gt_classes, pred_classes) if gt == pred)

# Calculate classification accuracy
if total_labels > 0:
    accuracy = (correct_predictions / total_labels) * 100
else:
    accuracy = 0.0

print(f"âœ… Script Classification Accuracy: {accuracy:.2f}%")
```

```python
import os
import glob
from PIL import Image

def xywhn_to_xyxy(x, y, w, h, img_w, img_h):
    # Convert normalized xywh to absolute xyxy format (xmin, ymin, xmax, ymax)
    xmin = (x - w / 2) * img_w
    ymin = (y - h / 2) * img_h
    xmax = (x + w / 2) * img_w
    ymax = (y + h / 2) * img_h
    return [xmin, ymin, xmax, ymax]

def iou(boxA, boxB):
    # box = [xmin, ymin, xmax, ymax]
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    if interArea == 0:
        return 0.0

    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])

    iou_val = interArea / float(boxAArea + boxBArea - interArea)
    return iou_val

# Paths
ROOT_DIR = '/content/gdrive/My Drive/add'
gt_dir = os.path.join(ROOT_DIR, 'yolo_formatted_dataset/labels/val')
pred_dir = '/content/runs/detect/val2/labels'  # update as per your validation output folder
img_dir = os.path.join(ROOT_DIR, 'yolo_formatted_dataset/images/val')  # update to your val images folder

# IoU threshold to consider a prediction correct
IOU_THRESHOLD = 0.5

gt_files = sorted(glob.glob(gt_dir + '/*.txt'))

total_gt = 0
total_correct = 0

for gt_file in gt_files:
    base_name = os.path.basename(gt_file)
    pred_file = os.path.join(pred_dir, base_name)
    img_file_jpg = os.path.join(img_dir, base_name.replace('.txt', '.jpg'))
    img_file_png = os.path.join(img_dir, base_name.replace('.txt', '.png'))

    # Get image size
    if os.path.exists(img_file_jpg):
        img = Image.open(img_file_jpg)
    elif os.path.exists(img_file_png):
        img = Image.open(img_file_png)
    else:
        print(f"Image file not found for {base_name}, skipping.")
        continue

    img_w, img_h = img.size

    # Read GT boxes and classes
    with open(gt_file, 'r') as f:
        gt_lines = f.readlines()

    gt_boxes = []
    for line in gt_lines:
        parts = line.strip().split()
        cls = parts[0]
        x, y, w, h = map(float, parts[1:5])
        box = xywhn_to_xyxy(x, y, w, h, img_w, img_h)
        gt_boxes.append({'class': cls, 'box': box, 'matched': False})

    total_gt += len(gt_boxes)

    # Read Predicted boxes and classes
    if not os.path.exists(pred_file):
        # No predictions for this image
        continue

    with open(pred_file, 'r') as f:
        pred_lines = f.readlines()

    pred_boxes = []
    for line in pred_lines:
        parts = line.strip().split()
        cls = parts[0]
        x, y, w, h = map(float, parts[1:5])
        box = xywhn_to_xyxy(x, y, w, h, img_w, img_h)
        pred_boxes.append({'class': cls, 'box': box, 'matched': False})

    # Match predictions to GT
    for pred in pred_boxes:
        best_iou = 0
        best_gt_idx = -1
        for i, gt in enumerate(gt_boxes):
            if gt['matched']:
                continue
            iou_val = iou(pred['box'], gt['box'])
            if iou_val > best_iou:
                best_iou = iou_val
                best_gt_idx = i

        if best_iou >= IOU_THRESHOLD:
            # Check class match
            if pred['class'] == gt_boxes[best_gt_idx]['class']:
                total_correct += 1
            gt_boxes[best_gt_idx]['matched'] = True

# Calculate accuracy
if total_gt > 0:
    accuracy = (total_correct / total_gt) * 100
else:
    accuracy = 0.0

print(f"âœ… Script Classification Accuracy (IoUâ‰¥{IOU_THRESHOLD}): {accuracy:.2f}%")
```
